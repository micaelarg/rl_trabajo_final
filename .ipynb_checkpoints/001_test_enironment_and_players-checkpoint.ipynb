# Importar entorno y familiarizarse

```python
from boardgame2 import ReversiEnv
import numpy as np
```

### Instanciación y reset

```python
board_shape = 8
env = ReversiEnv(board_shape=board_shape)
(board, player) = env.reset()
print(board)
print(player)
```

### Jugadas válidas

```python
env.get_valid((board, player))
```

### Pasos

```python
(board, player), reward, done, _ = env.step([2,4])
print(board, player, done)
```

# Implementar una partida completa

Muestreando el esapcio de acción con env.action_space.sample (uniformemente)

Notar que el estado es (board, player)

```python
done = False
(board, player) =  env.reset()
while not done: 
    action = env.action_space.sample() 
    print(board)
    print(f'action: {action}')
    
    print(f'player: {player}')
    (board, player), reward, done, _ = env.step(action)
    
```

Notar que muestrar uniformemente es un problema por que el episodio termina enseguida por la alta probabilidad de muestrar una acción no válida

# Implementar sample_valid_actions que muestrea uniformemente las acciones válidas

```python
env.get_valid((board,player))
```

```python
import random
```

```python
valid_actions = np.argwhere(env.get_valid((board,player)))
```

```python
def sample_valid_actions(state):
    valid_actions =  np.argwhere(env.get_valid((state)))
    max_bound = len(valid_actions)-1
    sampled_valid_action = valid_actions[random.randint(0, max_bound)]
    #print(type(sampled_valid_action))
    return sampled_valid_action
```

```python
# Verificar que funcione OK
sample_valid_actions((board, player))
```

```python
score_action = []
for action in valid_actions: 
    estado, recompensa, done, _ = env.next_step((board, player), action)
    #cuento la cantidad de fichas  
    score_action.append(abs(sum(list(estado[0].flatten()))))
```

### Correr ahora muestrando las acciones válidas y verificar que los epidosiod terminan en una situación donde ningún jugador puede mover mas

```python
done = False
(board, player) = env.reset()
while not done:
    action = sample_valid_actions((board, player))
    print(board)
    print(f'action: {action}')
    
    print(f'player: {player}')
    (board, player), reward, done, _ = env.step(action)
```
